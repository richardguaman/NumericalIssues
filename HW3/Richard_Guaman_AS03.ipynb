Calculate the Taylor Polynomials:
$$T_n (x) = \sum_{k=0}^{n} \frac{f^{(k)}(0)}{k!}(x)^k$$
centered at 0 for the following functions at the values of x,$$ x \in\ \{ \frac{1}{32}, \frac{1}{16}, \frac{1}{8}, \frac{1}{4}, \frac{1}{2}, 1, 2, 4 \} $$ 
and for
$$ n \in\ \{ 1, 2, 3, 4, 8, 16, 32, 64, 128, 256 \} $$

functions:

$$sin(x)$$
$$cos(x)$$
$$exp(x)$$
$$log_e(1-x)\textrm{   } x<1$$

```python
import math
from decimal import *
arr_x = [1/32, 1/16, 1/8, 1/4, 1/2, 1 ,2 ,4]
arr_n = [1, 2, 3, 4, 8, 16, 32, 64, 128, 256]
#For sin derivatives
sin_der = [0,1,0,-1]
#For cos derivatives
cos_der = [1,0,-1,0]
#For log(1-x) x<1 derivatives
log_der = [0, -1]
temp = 1
for i in range (1,256):
    log_der += [i*log_der[temp]]
    temp += 1

sin_sum_arr = []
cos_sum_arr = []
ex_sum_arr = []
log_sum_arr = []

true_sin = []
true_cos = []
true_ex = []
true_log = []

abs_err_sin = []
abs_err_cos = []
abs_err_ex = []
abs_err_log = []

rel_err_sin = []
rel_err_cos = []
rel_err_ex = []
rel_err_log = []

for i in arr_n:
    for j in arr_x:
        curr_sin_sum = 0
        curr_cos_sum = 0
        curr_ex_sum = 0
        curr_log_sum = 0
        temp = 0
        temp2 = 0
        for index in range (0,i+1):
            temp2 = (Decimal(math.pow(j,index)))/(Decimal(math.factorial(index)))
            curr_sin_sum += (sin_der[temp])*temp2
            curr_cos_sum += (cos_der[temp])*temp2
            #Derivative of exponential is itself. e^0 is always 1, so we set the exponential_sum equal to temp2.
            curr_ex_sum += temp2
            if(j<1):
                curr_log_sum += (log_der[index])*temp2
            temp += 1
            if temp == 4:
                temp = 0            
        sin_sum_arr += [curr_sin_sum]
        cos_sum_arr += [curr_cos_sum]
        ex_sum_arr += [curr_ex_sum]
        if(j<1):
            log_sum_arr += [curr_log_sum]
            
for i in arr_x:
    true_sin += [math.sin(i)]
    true_cos += [math.cos(i)]
    true_ex += [math.exp(i)]
    #Omitting the second argument of the log function automatically makes the base e.
    if(i<1):
        true_log += [math.log(1-i)]
temp = 0
for i in range(0,80):
    abs_err_sin += [abs(Decimal(true_sin[temp]) - sin_sum_arr[i])]
    abs_err_cos += [abs(Decimal(true_cos[temp]) - cos_sum_arr[i])]
    abs_err_ex += [abs(Decimal(true_ex[temp]) - ex_sum_arr[i])]
    rel_err_sin += [abs_err_sin[i]/Decimal(true_sin[temp])]
    rel_err_cos += [abs_err_cos[i]/Decimal(true_cos[temp])]
    rel_err_ex += [abs_err_ex[i]/Decimal(true_ex[temp])]
    temp += 1
    if temp % 8 == 0:
        temp = 0
temp = 0
for i in range(0,50):
    abs_err_log += [abs(Decimal(true_log[temp]) - log_sum_arr[i])]
    rel_err_log += [abs_err_log[i]/Decimal(true_log[temp])]
    temp += 1
    if temp % 5 == 0:
        temp = 0
```

```python
import pandas as pd
import numpy as np
 
#Create a DataFrame
d = {
    'Calculated Sum':[str(sin_sum_arr[0])],
    'Exact Value':[str(true_sin[0])],
    'Absolute Error':[str(abs_err_sin[0])],
    'Relative Error':[str(rel_err_sin[0])] 
}
print('For sin(x) when x = 1/32:') 
df = pd.DataFrame(d,columns=['Calculated Sum','Exact Value','Absolute Error','Relative Error'],index=['N = 1', 'N = 2', 'N = 3', 'N = 4', 'N = 8', 'N = 16','N = 32','N = 64', 'N = 128', 'N = 256'])
temp = 0
for row in df.index:
    for column in df.columns:
        if column == 'Exact Value':
            continue
        if column == 'Calculated Sum':
            df.at[row,column] = (sin_sum_arr[temp])
        if column == 'Absolute Error':
            df.at[row,column] = (abs_err_sin[temp])
        if column == 'Relative Error':
            df.at[row,column] = (rel_err_sin[temp])
    temp = temp + 8
df
```

As seen from the results above, as "N", the number of terms used in the Taylor Polynomial summation, increases, the closer the calculated results are to the actual results. Another interesting observation that can be made is that if the x term increases, getting further from the point at which the Taylor Polynomial is centered, the less accurate the results become with a lower number of "N" terms. For example, when x = 4 and N = 1 for exp(x), the calculated and exact values differ by a sizable margin. This is due to the fact that the Taylor Polynomial is centered at 0, so x values further from 0 will need more "N" values to approach the actual value of the function at that x value.
