```python
## Method 1: Lagrange Interpolating Polynomial
import math, pandas as pd
from math import sqrt

def lagrange(x_val,y_val,x):
    running_sum = 0
    for i in range(0, len(y_val)):
        numerator = 1
        denominator = 1
        for j in range(0,len(x_val)):
            if(i == j):
                continue 
            else:
                numerator = numerator * (x - x_val[j])
                denominator = denominator*(x_val[i] - x_val[j])            
        running_sum = running_sum + y_val[i] * numerator /denominator
    return running_sum

data = pd.read_excel(r'C:\Users\Richard\Desktop\Numerical Issues\Final Project\data.xlsx')
time = pd.DataFrame(data,columns=['t'])
accel = pd.DataFrame(data,columns=['a'])
x_val = []
y_val = []
actual_results = []
lagrange_results = []
#print(years.iloc[0,0])
for i in range(28,53,2):
    x_val += [time.iloc[i,0]]
    y_val += [accel.iloc[i,0]]
#for i in range(0,13):
    #print(x_val[i])
temp = 1.5
for i in range(0,12):
    lagrange_results += [lagrange(x_val,y_val,temp)]
    #print(lagrange(x_val,y_val,temp))
    temp += .1
print("")

for i in range(29,53,2):
    actual_results += [accel.iloc[i,0]]

    #print(accel.iloc[i,0])
#print(lagrange(x_val,y_val,2.40))
```

```python
# Linear Interpolation
def pli(x_val,y_val,target):
    approximation = 0
    if(target < x_val[0] or target >  x_val[len(x_val)-1]):
        print('The target x value must be between the given x domain.')
        return
    if(target == x_val[0]):
        return y_val[0]
    if(target == x_val[len(x_val)-1]):
        return y_val[len(y_val)]
    for i in range (0, len(x_val)):
        if(target >= x_val[i] and target <= x_val[i+1]):
            slope = 0
            b = 0
            slope = (y_val[i+1] - y_val[i])/ (x_val[i+1]-x_val[i])
            b = y_val[i] - (slope*x_val[i])
            approximation = slope*target + b
            return approximation
        
temp = 1.5
linear_results = []
for i in range(0,12):
    linear_results += [pli(x_val,y_val,temp)]
    #print(pli(x_val,y_val,temp))
    temp += .1
#print("")    
#for i in range(29,53,2):
#    print(accel.iloc[i,0])
```

```python
# Method 3: Netwon's Divided Difference Method

def ndd(x_val, y_val, target):
    calculated_sum = 0
    temp_var = len(x_val) - 1
    temp = 0
    num_arr = []
    values = []
    for i in range(0, len(y_val)):
        num_arr += [y_val[i]]
    for i in range (0, len(x_val)-1):
        temp2 = i + 1
        for j in range (0, temp_var):
            num_arr.append((num_arr[temp + 1] - num_arr[temp])/(x_val[temp2] - x_val[j]))
            values.append((num_arr[temp + 1] - num_arr[temp])/(x_val[temp2] - x_val[j]))
            temp = temp + 1
            temp2 = temp2 + 1
        temp = temp + 1
        temp_var = temp_var - 1
    
    temp_product = 1
    temp2 = 2
    temp = len(values) - 1
    temp_var = len(x_val) - 1 
    for i in range (0, len(x_val)-1): 
        for j in range (0, temp_var):
            temp_product = temp_product * (target - x_val[j])
        calculated_sum = calculated_sum + values[temp] * temp_product
        temp = temp - temp2
        temp2 = temp2 + 1
        temp_product = 1
        temp_var = temp_var - 1
    calculated_sum = calculated_sum + y_val[0]
    return calculated_sum

temp = 1.5
newton_results = []
for i in range(0,12):
    newton_results += [ndd(x_val,y_val,temp)]
    #print(ndd(x_val,y_val,temp))
    temp += .1
#print("")    
#for i in range(29,53,2):
#    print(accel.iloc[i,0])
```

```python
# Method 4: Neville's Method
def nev(x_val, y_val, target):
    calculated_sum = 0
    temp = 1
    coef_arr = []
    for i in range(0, len(y_val)):
        coef_arr += [y_val[i]]
    arr_len = len(x_val)
    res_len = arr_len - 1
    for i in range (arr_len, 0, -1):
        for j in range (0, res_len):
            coef_arr[j] = ((target - x_val[j+temp])*coef_arr[j] - (target - x_val[j])*coef_arr[j+1])/(x_val[j] - x_val[j+temp])        
        res_len = res_len - 1
        temp = temp + 1
    return(coef_arr[0])

temp = 1.5
neville_results = []
for i in range(0,12):
    neville_results += [nev(x_val,y_val,temp)]
    #print(nev(x_val,y_val,temp))
    temp += .1
#print("")    
#for i in range(29,53,2):
#    print(accel.iloc[i,0])
```
